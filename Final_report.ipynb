## 3. Methodology

### 3.1 Gathering and processing the data corresponding to the location under study
1. The perliminary step to begin this analysis is to capture all the necessary data related to the location that we take for the study. The location considered in this case is Toronto, Canada.
2. The first required data is the list of neighborhoods in and around Toronto. The data is obtained from the wikipedia link https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M (oldid: 954731857).
3. BeautifulSoup library is used to scrap the data from the above mentioned link. A dataframe is created out of the resulted data. The raw data is processed and cleaned to get a more simplified informtion that is analysis ready.
4. The second most important data required is the coordinates for each of the neighbourhoods in Toronto. The geo coordinates were taken from the readily available geospatial csv from the following link - https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DS0701EN-SkillsNetwork/labs_v1/Geospatial_Coordinates.csv"
5. Both these two seperate data were matched based on the postal code and a final dataframe is created, which primarily includes the neighborhood names, postal codes, latitudes and longitudes

### 3.2 Collect information about the existing laudries and accomodation facilities
1. Foursquare API will be used to query all the listed neighborhoods of central Toronto for the currently active hotels, motels, laudromats.
2. The latitudes and logitudes from the earlier created dataframe will be used as an input for these queries.
3. The result will be stored into new dataframes. 
4. Folium will be used to visualise each of these venues in the map for better visual understanding.
5. Heirarchial clustering will be performed on the laudry and hotel data to identify the concentration of each of the venues in the neighborhoods.
6. The final clustering data will be plotted using folium. This gives a better understanding of areas with adequate or scarce establishment of hotels and laudromats.
7. For our conclusion purpose, the areas which has more hotels and less laudromats will be suggested for establishing new laudry services. 


## 4. Results
1. A total of 9 neighborhoods were identified within central Toronto.
2. A total of 111 laudromat venues were listed within central Toronto.
3. A total of 90 Hotel venues were listed within central Toronto.
4. Based on Heirarchial clustering, 6 clusters of Laudromats and 4 clusters of hotels were generated.
5. The laudromat to Hotels ratio is very less in the ForestHill neighborhood. It has got 6 hotels with no associated laudry services.

## 5. Discussions
1. As assumed earlier, in most of the neighborhoods, the hotel clusters and the laudromat clusters overlap each other. 
2. Majority of the hotels are found in and around the Yorkville neighborhood. The main reason could be the presence of the museum and university around this neighborhood. We could also see a complimenting numbers of laudromats around this neighborhood.
3. The annex neighborhood got the ratio of laudromat to hotel very high as it turns out to be more of a residential area along with schools, which raises a demand for laudry services inspite of having lesser hotels.
4. The Forest Hill neighborhood got a small cluster of hotels but does not have any laudry establishments to meet the need.

## 6. Conclusions
Based on the above results and discussions, we shall come to a conclusion that the Foresthill neighborhood is the most suitable candidate for launching a new laundromat services, which might satify the demand and stands for a profitable business establishment.
